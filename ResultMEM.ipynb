```python
import pandas as pd
import matplotlib.pyplot as plt

smote_rf_Recall = 0.831081
smote_rf_Precision = 0.580189
smote_rf_f1 = 0.683333
smote_rf_accuracy = 0.998666

smote_ann_Recall = 0.972973
smote_ann_Precision = 0.006737
smote_ann_f1 = 0.013381
smote_ann_accuracy = 0.751472

smote_XGBoost_Recall = 0.864865
smote_XGBoost_Precision = 0.159006
smote_XGBoost_f1 = 0.268625
smote_XGBoost_accuracy = 0.991843

smote_AdaBoost_Recall = 0.858108
smote_AdaBoost_Precision = 0.103758
smote_AdaBoost_f1 = 0.185131
smote_AdaBoost_accuracy = 0.986915

smote_scores = [
    ("Random Forest", smote_rf_Recall, smote_rf_Precision, smote_rf_f1, smote_rf_accuracy),
    ("ANN", smote_ann_Recall, smote_ann_Precision, smote_ann_f1, smote_ann_accuracy),
    ("XGBoost", smote_XGBoost_Recall, smote_XGBoost_Precision, smote_XGBoost_f1, smote_XGBoost_accuracy),
    ("AdaBoost", smote_AdaBoost_Recall, smote_AdaBoost_Precision, smote_AdaBoost_f1, smote_AdaBoost_accuracy)
]

smote_score_df = pd.DataFrame(smote_scores, columns=['Model', 'Recall', 'Precision', 'F1 Score', 'Accuracy'])

smote_score_df['Sampling Method'] = 'SMOTE'

print(smote_score_df)


plt.figure(figsize=(10, 6))
plt.axis('tight')
plt.axis('off')
the_table = plt.table(cellText=smote_score_df.values, colLabels=smote_score_df.columns, cellLoc='center', loc='center')


the_table.auto_set_font_size(False)
the_table.set_fontsize(12)
the_table.scale(1.2, 1.2)


image_filename = 'smote_scores.png'
plt.savefig(image_filename, bbox_inches='tight', pad_inches=0.5)
plt.close()

print(f"Tableau exporté sous forme d'image : {image_filename}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Vos données
smote_rf_auc = 0.9150
smote_ann_auc = 0.8620
smote_XGBoost_auc = 0.9285
smote_AdaBoost_auc = 0.9226

smote_auc_scores = [
    ("Random Forest", smote_rf_auc),
    ("ANN", smote_ann_auc),
    ("XGBoost", smote_XGBoost_auc),
    ("AdaBoost", smote_AdaBoost_auc)
]


smote_auc_df = pd.DataFrame(smote_auc_scores, columns=['Model', 'AUC-ROC'])


smote_auc_df['Sampling Method'] = 'SMOTE'

smote_auc_df = smote_auc_df.sort_values(by='AUC-ROC', ascending=False)


plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='AUC-ROC', data=smote_auc_df, palette='viridis')
plt.title('AUC-ROC Scores for Different Models under SMOTE Sampling')
plt.xlabel('Model')
plt.ylabel('AUC-ROC Score')
plt.ylim(0.8, 0.95) 
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()
```
